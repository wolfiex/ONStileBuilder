```python
import geopandas as gpd
```

```python
df = gpd.read_file('./data/Counties_and_Unitary_Authorities_(December_2011)_Boundaries_EW_BGC.geojson')[['ctyua11cd','ctyua11nm','geometry']]
```

```python
newdf = df.merge(df.bounds,left_index=True,right_index=True)
```

```python
newdf
```

```python
newdf.to_file("./tiles/CountiesUA.geojson", driver="GeoJSON")
```
